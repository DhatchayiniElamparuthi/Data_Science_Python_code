```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#matplotlib to render plots in the notebook
%matplotlib inline
```

```python
import warnings
warnings.simplefilter('ignore')
```

```python
tips=sns.load_dataset('tips') #importing the inbuilt dataset "tips"
```

```python
tips
```

### Strip plot

```python
sns.stripplot(y="tip", data=tips, jitter=True)
plt.ylabel("tips($)")
plt.title("Strip Plot", size=20)
```

```python
#grouping with stripplot()- here one variable must be discrete and the other must be continuous
sns.stripplot(x="day", y="tip", data=tips) #by default, jitter is true
plt.ylabel("tip($)")
plt.title("Strip Plot for 2 variables", size=20)
```

#### From the above stripplot, we can interpret on which day maximum visitors gave tips(saturday) and similarly on which day minimum visitors gave tips(friday)

### Swarmplot

```python
sns.swarmplot(x="day", y="tip", data=tips)
plt.ylabel("tip($)")
plt.title("Swarm Plot", size=20)
```

#### From the above swarmplot also, we can say, on which day max/min no. of visitors have given tips. In addition to this, we can also say on which day max no. of people have given tip as 3 dollar(which is  saturday); similarly how many people have given tip as 3 dollar(which is 4 people), like these kind of insights can be drawn

```python
sns.swarmplot(x="day", y="tip", data=tips, hue="size") #here hue variable must be discrete
plt.ylabel("tip($)")
plt.title("Swarm Plot for the discrete column 'size' wrt tips and day", size=20)
```

#### From the above swarmplot, we can say that most of the vistors are of size 2(i.e pair) and rarely it is of size 6.

```python
#trying the same above plot for discrete column "sex"
sns.swarmplot(x="day", y="tip", data=tips, hue="sex") #here hue variable must be discrete
plt.ylabel("tip($)")
plt.title("Swarm Plot for the discrete column 'sex' wrt tips and day", size=20)
```

#### From the above swarmplot, the insights that could be drawn are : on which day max. females have given tips(thursday), who has given tip as 10 dollar(male), who has given min. tips(females) and so on...

```python
#trying the same above plot for discrete column "time"
sns.swarmplot(x="day", y="tip", data=tips, hue="time") #here hue variable must be discrete
plt.ylabel("tip($)")
plt.title("Swarm Plot for the discrete column 'time' wrt tips and day", size=20)
```

#### From the above swarmplot, the insights that could be drawn are: mostly on weekends, people come only for dinner(so, accordingly on weekends, the hotel can give some dicounts during lunch time); and on thursdays, people mostly come for lunch, and so on..

```python
#trying the same above plot for discrete column "time"
sns.swarmplot(x="day", y="tip", data=tips, hue="day") #here hue variable must be discrete
plt.ylabel("tip($)")
plt.title("Swarm Plot for the discrete column 'day' wrt tips and day", size=20)
```

```python
#trying the same above plot for discrete column "smoker"
sns.swarmplot(x="day", y="tip", data=tips, hue="smoker") #here hue variable must be discrete
plt.ylabel("tip($)")
plt.title("Swarm Plot for the discrete column 'smoker' wrt tips and day", size=20)
```

#### From the above swarmplot, we can interpret that, most of the visitors are non-smokers

```python
#horizontal swarm plot can also be plotted by giving orientation as horizontal(orient= 'h')
sns.swarmplot(x="tip", y="day", data=tips, hue="sex", orient='h') #remember to change the x and y axis
plt.ylabel("tip($)")
plt.title("Horizontal swarm plot", size=20)
```

### Box and Violin plot

```python
sns.boxplot(x="day", y="tip", data=tips)
plt.ylabel("tip($)")
plt.title("Box plot", size=20)
```

#### From the above plot, we can interpret that max. no of outliers is on saturday. Apart from the outliers, we can also interpret, on which day max. tip is given(sunday), on which days min. tip is given(friday, saturday, sunday)

```python
sns.violinplot(x="day", y="tip", data=tips)
plt.ylabel("tip($)")
plt.title("Violin plot", size=20)
```

#### If we rotate the above plot clockwise,then we can see that saturday is highly positively skewed and friday is somewhat a symmetric distribution.Other insights that can be drawn are: on thursday, max no. of people have given tip in b/w 1 to 3 dollar, similarly on friday, max no. of people have given tip in b/w 3 to 4 dollar, and so on. So, wherever it is bulky, max. no of data points lie in that particular region.
#### The black rod like structure in the above violin plots, is called confidence interval.

```python
#combining the above boxplot and violin plot as subplots in a single figure
plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')

plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')
```

```python
#combining 2 plots one over the other(here combining violin plot and strip plot)
sns.violinplot(x='day', y='tip', data=tips, inner=None, color='lightgray') #inner=None-->confidence interval bar is removed
sns.stripplot(x='day',y='tip',data=tips, size=4, jitter=True)#size=4--->represents the dot size
plt.ylabel('tip($)')
```

```python
#Joint plot
sns.jointplot(x='total_bill', y='tip', data=tips)
```

#### Here 2 plots have been created, inside it is scatter plot b/w total_bill & tip , whereas outside it is histogram (as both the 'total_bill' & 'tip' columns are continuous, histogram is generated)
#### The insights that can be drawn are: From the scatter plot, we can interpret that as the 'total_bill' increases, the 'tip' also increases. From the histogram, the 'total_bill' is right skewed(as tail is towards right side), whereas the histogram for 'tip' is left skewed(as tail is towards left side)

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

#### This above plot is also giving the same insights as that of before overlapped plot(i.e scatterplot and histogram).Here all the 3 plots are density plots.

```python
#pairplot
sns.pairplot(tips)
```

#### In the above pairplot, diagonally histogram/barplot is generated depending on whether it is discrete('size') or continuous(here 'total_bill', 'tip') and in remaning areas scatter plot is generated.

```python
sns.pairplot(tips, hue='day', diag_kws={'bw':0.9}) #value for hue variable must be a discrete column
```

```python
sns.pairplot(tips, hue='sex', diag_kws={'bw':0.9})
```

```python
sns.pairplot(tips, hue='size', diag_kws={'bw':0.9})
```

#### In the above 3 pairplots, we have used a third variable 'hue', where diagonally density plots is created and in remaining areas scatter plot is generated.
