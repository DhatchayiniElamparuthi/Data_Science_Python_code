## Visualization in pandas

```python
import pandas as pd
df=pd.DataFrame(index=['Riya','Smriti','John','Dhipa','Tarun'],
               data={'Apples':[20,10,40,20,50], 'Oranges':[35,40,25,19,33]}) #creating a small dataframe
```

```python
df
```

```python
#barplot
a=df.plot.bar(color=['0','0.4'], figsize=(16,4)) #visualization in pandas gives color ranging from white to black, where white-1,black-0
```

```python
#density plot
a=df.plot.kde(color=['0.1','0.6'],figsize=(16,4))
```

#### From the above density plot,we can see that apples have flat curve, which means negative kurtosis, that means it has high variance. And for oranges have sharp curve (when compared to apples), which means positive kurtosis, that means low variance.

### Therefore,visualization in pandas, provides a color code which only ranges from white to black. So, visualization is very basic in pandas. Hence, inorder to have advanced visualization, seaborn is used. 

```python
#plotting lineplot,scatter plot,barplot in a single figure using subplots function from matplotlib
fig,(ax1,ax2,ax3)=plt.subplots(1,3, figsize=(16,4))
fig.suptitle("Two Variable plots",size=15)
df.plot.line(ax=ax1, title='Line plot')
df.plot.scatter(x='Apples', y='Oranges', ax=ax2, title='Scatter plot')
df.plot.bar(ax=ax3, title='Bar plot')
```

```python
fig,(ax1,ax2,ax3)=plt.subplots(1,3, figsize=(16,4))
fig.suptitle("Two Variable plots",size=15)
df.plot.kde(ax=ax1, title='KDE plot')
df.plot.box(ax=ax2, title="Boxplot")
df.plot.hist(ax=ax3, title='Histogram')
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2, figsize=(14,7))
df.sort_values('Apples').plot.line(x='Apples', y='Oranges',ax=ax1)#line plot for oranges wrt to apples
df.plot.bar(x='Apples', y='Oranges', ax=ax2, color='green') #plotting barplot for oranges wrt to apples
df.plot.kde(x='Oranges',ax=ax3)#kde plot for apples wrt to oranges
df.sort_values('Oranges').plot.line(x='Oranges',y='Apples',ax=ax4)#line plot for apples wrt to oranges
```
