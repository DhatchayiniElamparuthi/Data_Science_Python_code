### groupby method and performing some operations on it 

```python
#groupby() can only be applied on discrete values containing columns
abc=df.groupby(['sex'])#here 2 grps will be created, since it is bimodal
```

```python
abc.mean()
```

```python
grouped=df.groupby('sex')
```

```python
grouped.groups #the respective indices of the grouped column is displayed
```

```python
abc=df.groupby(['rank'])#or df.groupby(['rank']).mean()
```

```python
abc.mean()
```

```python
df.groupby('rank')[['salary']].mean()#if we use single sq bracket for salary,then the same output will be in a non-tabular format
```

```python
df.groupby('rank')[['salary','phd']].mean()#for multiple columns,we should use double sq brackets else warning comes
                                           #for single column,like the 'salary' in above cell,we can either use single or double sq brackets
```

### Slicing the pandas data frame using dataframe.iloc[]

##### iloc function[rows,columns]--->here in this dataset,it starts from 0th column,so 0th column will be column1 
##### 1st column-->column2
##### 2nd column--->column 3 and so on... similarly, for the rows

```python
df.iloc[0] #0th row data
```

```python
df.iloc[15] #15th row data
```

```python
df.iloc[-1] #last row data
```

```python
df.iloc[:,2] #all the  rows of 2nd column
```

```python
df.iloc[10:20,] #10th row to 19th row and all the columns
```

```python
df.iloc[:,-1] #last column
```

```python
df.iloc[0:4,0:3] #0th to 3rd row,0th to 2nd column 
```

```python
df.iloc[1:4,2:5] #1st to 3rd row ; 2nd to 4th column
```

```python
#to display "particular" data,use comma instead of colon
df.iloc[[1,5],[3,5]] #particulary 1st,5th row and 3rd,5th column will be displayed
```
